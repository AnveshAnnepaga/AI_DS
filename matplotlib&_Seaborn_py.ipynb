```python
# task1
print("Shape of dataset:", df.shape)
print("\nColumn data types:")
print(df.dtypes)


print("\nMissing values per column:")
print(df.isnull().sum())

# task2
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)

# task3
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Count invalid or missing dates (turned into NaT)
invalid_dates = df['Date'].isna().sum()
print(f"\nInvalid or missing dates: {invalid_dates}")

# Sort by date
df = df.sort_values(by='Date')

# task4
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numeric columns:", list(numeric_cols))
for col in numeric_cols:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)

print("\nMissing values after imputation:")
print(df.isnull().sum())
```

```python
plt.plot(df["Calories"],df["Date"],"o--r")
plt.grid(True)
plt.title("calories vs date")
plt.xlabel("calories")
plt.ylabel("date")
plt.savefig("line_calories_over_time.png")
plt.show()
```

```python

bins = [0, 30, 60, 90, 120, 150, 180]
labels = ["0-30", "31-60", "61-90", "91-120", "121-150", "151-180"]

# Create a new column for duration groups
df["Duration_Group"] = pd.cut(df["Duration"], bins=bins, labels=labels, include_lowest=True)

# Compute average Pulse per duration group
avg_pulse = df.groupby("Duration_Group")["Pulse"].mean()

# Plot bar chart
avg_pulse.plot(kind="bar", color="skyblue", edgecolor="black")

plt.title("Average Pulse by Duration Group")
plt.xlabel("Duration (minutes)")
plt.ylabel("Average Pulse")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Save the chart
plt.savefig("bar_avg_pulse_by_duration.png")
plt.show()
```

```python
from scipy.stats import pearsonr

#Duration vs Calories
plt.figure(figsize=(7,5))
plt.scatter(df['Duration'], df['Calories'], color='blue', alpha=0.6, label='Data points')

# Linear trend line
m, b = np.polyfit(df['Duration'], df['Calories'], 1)
plt.plot(df['Duration'], m*df['Duration'] + b, color='red', linestyle='--', label='Trend line')

#Pearson correlation
corr, _ = pearsonr(df['Duration'].dropna(), df['Calories'].dropna())
plt.title(f'Duration vs Calories (r = {corr:.2f})')
plt.xlabel('Duration (minutes)')
plt.ylabel('Calories')
plt.legend()
plt.grid(True)
plt.show()


#pie chart
bins = [0, 120, 140, 160, df['Maxpulse'].max()]
labels = ['<=120', '121-140', '141-160', '>160']
df['MaxpulseRange'] = pd.cut(df['Maxpulse'], bins=bins, labels=labels, include_lowest=True)
counts = df['MaxpulseRange'].value_counts().sort_index()

plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=['lightblue','lightgreen','orange','red'], explode=[0.05]*4)
plt.title("Maxpulse Distribution")
plt.savefig("/content/drive/MyDrive/pie_maxpulse_ranges.png", bbox_inches='tight')
plt.show()


#hist
plt.figure(figsize=(7,5))
plt.hist(df['Calories'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram of Calories")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig("/content/drive/MyDrive/hist_calories.png", bbox_inches='tight')
plt.show()

# Skewness comment
skewness = df['Calories'].skew()
print(f"Calories histogram skewness: {skewness:.2f}")
if skewness > 0:
    print("Distribution is positively skewed (long tail on the right).")
elif skewness < 0:
    print("Distribution is negatively skewed (long tail on the left).")
else:
    print("Distribution is roughly symmetric.")


#multi lone scatter

plt.figure(figsize=(14,5))
# Left: Pulse over Date
plt.subplot(1,2,1)
plt.plot(df['Date'], df['Pulse'], 'o-', color='purple', label='Pulse')
plt.title("Pulse Over Time")
plt.xlabel("Date")
plt.ylabel("Pulse")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# Right: Duration vs Maxpulse
plt.subplot(1,2,2)
plt.scatter(df['Duration'], df['Maxpulse'], color='teal', alpha=0.6)
plt.title("Duration vs Maxpulse")
plt.xlabel("Duration (minutes)")
plt.ylabel("Maxpulse")
plt.grid(True)

plt.tight_layout()
plt.savefig("/content/drive/MyDrive/subplots_pulse_duration_maxpulse.png", bbox_inches='tight')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')

# Load example datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')

print('Datasets loaded: tips (rows={}), iris (rows={})'.format(len(tips), len(iris)))

tips.head()
sns.histplot(tips['total_bill'], kde=False)
plt.title('Histogram - Distribution of Total Bill')
plt.show()
# KDE Plot
sns.kdeplot(tips['total_bill'])
plt.title('KDE Plot - Smooth Distribution of Total Bill')
plt.show()
```
