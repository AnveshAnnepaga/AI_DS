Indexing and Slicing:
You are working as a Data Analyst at a fruit distribution company.
The company stores the number of fruit boxes shipped per day (for 1 week) in a 2D NumPy array where:
Each row represents a type of fruit.
Each column represents a day of the week (Mon–Sun).


Your Tasks
Create the 2D NumPy array
Create an array named shipments using the above data.
Access single elements (Indexing):
Get the number of Apples shipped on Wednesday.
Get the number of Oranges shipped on Sunday.
Access multiple elements (Slicing):
Extract all the weekday shipments (Mon–Fri) for all fruits.
Extract only the weekend shipments (Sat & Sun).
Extract the shipments of Apples from Wednesday to Saturday.
Bonus (Practice thinking):
Replace the Saturday shipment of Apples (currently 80) with 100 using indexing.

```python
import numpy as np

shipments = np.array([[50, 60, 55, 70, 65, 80, 90],
                      [40, 45, 50, 60, 55, 70, 75]])

print("apples on wed", shipments[0, 2])
print("oranges on sunday:",shipments[1, 6])

print("mod-fri:",shipments[0:, 0:5])
print("sat and sun:",shipments[0:, 5:])
print("wed to sat(apple)", shipments[0, 2:6])

shipments[0, 5] = 100
print(shipments)
```

```python
farmA = np.array([[10,12,11],[9,11,10],[12,13,14],[8,7,9]])
farmB = np.array([[20,19,21],[18,20,19],[22,23,21],[17,18,16]])
farmC = np.array([[15,16,14],[14,13,15],[16,17,18],[12,11,13]])
production_all = np.array([farmA, farmB, farmC])

print(production_all)
print("shape:", production_all.shape)
print("size :", production_all.size)
print("data type:", production_all.dtype)
print("dimensions:", production_all.ndim)

total_per_farm = np.sum(production_all, axis=(1,2))
mean_per_farm = np.mean(production_all, axis=(1,2))
std_per_farm = np.std(production_all, axis=(1,2))
var_per_farm = np.var(production_all, axis=(1,2))

print(total_per_farm)
print(mean_per_farm)
print(std_per_farm)
print(var_per_farm)

hour_weights = np.array([0.5, 1.0, 0.75])
print("weighted_per_panel:", np.dot(production_all, hour_weights))

production_all[1] = production_all[1] + 0.5

normalized = (production_all - mean_per_farm[:, None, None]) / std_per_farm[:, None, None]

exp_vals = np.exp(normalized)
expm1_vals = np.expm1(normalized)
log_vals = np.log(normalized - np.min(normalized) + 1e-5)
log10_vals = np.log10(normalized - np.min(normalized) + 1e-5)
log2_vals = np.log2(normalized - np.min(normalized) + 1e-5)

sin_vals = np.sin(normalized)
cos_vals = np.cos(normalized)
tan_vals = np.tan(normalized)
arcsin_vals = np.arcsin(np.clip(normalized, -1, 1))
arccos_vals = np.arccos(np.clip(normalized, -1, 1))
arctan_vals = np.arctan(normalized)
degrees_vals = np.degrees(normalized)
radians_vals = np.radians(normalized)

round_vals = np.round(normalized)
floor_vals = np.floor(normalized)
ceil_vals = np.ceil(normalized)
trunc_vals = np.trunc(normalized)

high_output = production_all > 20
even_hours_mask = (np.arange(3) % 2 == 0)
logical_and_vals = np.logical_and(high_output, high_output)
logical_or_vals = np.logical_or(high_output, high_output)
logical_not_vals = np.logical_not(high_output)
logical_xor_vals = np.logical_xor(high_output, high_output)
greater_equal_vals = np.greater_equal(production_all, 15)
less_vals = np.less(production_all, 10)

alerts = np.array([[1,0,1],[0,0,1],[1,1,0],[0,1,0]])
bitwise_and_vals = np.bitwise_and(alerts, alerts)
bitwise_or_vals = np.bitwise_or(alerts, alerts)
bitwise_invert_vals = np.invert(alerts)

add_vals = np.add(normalized, 1)
sub_vals = np.subtract(normalized, 1)
mul_vals = np.multiply(normalized, 2)
div_vals = np.divide(normalized, 2)
power_vals = np.power(normalized, 2)
mod_vals = np.mod(normalized, 2)

sum_exp = np.sum(np.exp(normalized))
mean_sin = np.mean(np.sin(normalized))

panel_totals_per_farm = np.sum(production_all, axis=2)
needs_maintenance = np.any(panel_totals_per_farm < 40, axis=1)

overall_total = np.sum(production_all)
peak_hour_value = np.max(production_all)
peak_idx = np.unravel_index(np.argmax(production_all), production_all.shape)
average_per_panel = np.mean(production_all, axis=2)

print("Total per farm:", total_per_farm)
print("Mean per farm:", mean_per_farm)
print("Std per farm:", std_per_farm)
print("Panel totals per farm:\n", panel_totals_per_farm)
print("Needs maintenance:", needs_maintenance)
print("Overall total:", overall_total)
print("Peak value:", peak_hour_value, "at farm, panel, hour:", peak_idx)
print("Average per panel:\n", average_per_panel)
```
