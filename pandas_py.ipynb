```python
import pandas as pd
import numpy as np

# Load dataset
file_path = '/content/drive/MyDrive/OnlineRetail.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Inspect
print("Head of dataset")
print(df.head())

print("Tail of dataset")
print(df.tail())

print("Dataset info")
print(df.info())

print("Shape of dataset")
print(df.shape)

print("Data types")
print(df.dtypes)

# Missing Data Handling
print("Missing values per column")
print(df.isnull().sum())

df = df.dropna(subset=['CustomerID'])
df['Description'] = df['Description'].fillna('Unknown')

# Date parsing
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df = df.dropna(subset=['InvoiceDate'])

# Outliers
returns = df[df['Quantity'] < 0]
print("Number of returns")
print(returns.shape[0])

df = df[df['UnitPrice'] > 0]

# Duplicates
duplicates = df.duplicated()
print("Number of duplicates")
print(duplicates.sum())
df = df.drop_duplicates()
df = df.reset_index(drop=True)

print("Summary statistics")
print(df.describe())

print("Dataset info after cleaning")
print(df.info())

# Filtering & Sorting
print("Transactions with Quantity > 1000")
suspicious_outliers = df[df['Quantity'] > 1000]
print(suspicious_outliers)

print("Top revenue orders")
df['Revenue'] = df['Quantity'] * df['UnitPrice']
sorted_revenue = df.sort_values(by='Revenue', ascending=False)
print(sorted_revenue.head())

# Merge / Enrich Data
print("Merge with product_info")
product_info = pd.DataFrame({
    'StockCode': ['10002', '10003', '10004'],
    'Category': ['Stationery', 'Decor', 'Electronics'],
    'Supplier': ['Supplier A', 'Supplier B', 'Supplier C']
})
merged_df = pd.merge(df, product_info, on='StockCode', how='left')
print(merged_df.head())

# Grouping & Aggregation
print("Revenue by Country")
country_stats = merged_df.groupby('Country').agg(
    Total_Revenue=('Revenue', 'sum'),
    Mean_Revenue=('Revenue', 'mean'),
    Order_Count=('InvoiceNo', 'nunique')
).reset_index()
print(country_stats)

print("Price range summary")
bins = [0, 5, 20, np.inf]
labels = ['Cheap', 'Medium', 'Expensive']
merged_df['Price_Range'] = pd.cut(merged_df['UnitPrice'], bins=bins, labels=labels)
price_summary = merged_df.groupby('Price_Range').agg(
    Purchases=('InvoiceNo', 'count'),
    Avg_Revenue=('Revenue', 'mean')
).reset_index()
print(price_summary)

# Correlation
print("Correlation among numeric columns")
correlation = merged_df[['Quantity', 'UnitPrice', 'Revenue']].corr()
print(correlation)

# Rename & String Operations
print("Rename columns")
merged_df.rename(columns={'StockCode': 'Product_ID', 'UnitPrice': 'Price'}, inplace=True)
print(merged_df.head())

print("Products containing 'glass'")
merged_df['Description'] = merged_df['Description'].str.lower()
glass_products = merged_df[merged_df['Description'].str.contains('glass', na=False)]
print(glass_products.head())

# Descriptive Statistics
print("Descriptive statistics for numeric columns")
print(merged_df.describe())

print("Value counts for Country")
print(merged_df['Country'].value_counts())

print("Value counts for Category")
print(merged_df['Category'].value_counts(dropna=False))

# Advanced Filtering / Querying
print("Orders where Revenue > 1000 and Country == 'Germany'")
high_value_germany = merged_df.query("Revenue > 1000 and Country == 'Germany'")
print(high_value_germany)

print("Transactions where Quantity between 10 and 100")
quantity_range = merged_df[merged_df['Quantity'].between(10, 100)]
print(quantity_range.head())

# Export Cleaned & Enriched Data
print("Saving final cleaned and enriched dataset")
final_path_csv = '/content/drive/MyDrive/OnlineRetail_Final.csv'
final_path_excel = '/content/drive/MyDrive/OnlineRetail_Final.xlsx'
merged_df.to_csv(final_path_csv, index=False)
merged_df.to_excel(final_path_excel, index=False)
print("Final dataset saved successfully to CSV and Excel")
```
